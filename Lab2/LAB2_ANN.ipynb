```python
# Import cГЎc thЖ° viб»‡n cбє§n thiбєїt dГ№ng chung cho cГЎc bГ i tбє­p
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, Flatten, Dropout
import matplotlib.pyplot as plt
```

BГЂI 1

```python
# BГ i 1: ANN cho bб»™ dб»Ї liб»‡u CIFAR10
print("--- BГ i 1: CIFAR10 ---")

# 1. TбєЈi dб»Ї liб»‡u
cifar10 = tf.keras.datasets.cifar10
(X_train_c10, y_train_c10), (X_test_c10, y_test_c10) = cifar10.load_data()

# 2. Tiб»Ѓn xб»­ lГЅ dб»Ї liб»‡u (Chuбє©n hГіa pixel vб»Ѓ khoбєЈng 0-1)
X_train_c10 = X_train_c10 / 255.0
X_test_c10 = X_test_c10 / 255.0

# 3. Khб»џi tбєЎo mГґ hГ¬nh ANN
model_1 = Sequential([
    Flatten(input_shape=(32, 32, 3)), # Chuyб»ѓn бєЈnh 3D (32x32x3) thГ nh 1D
    Dense(512, activation='relu'),
    Dropout(0.2),
    Dense(256, activation='relu'),
    Dense(10, activation='softmax')
])

# 4. BiГЄn dб»‹ch mГґ hГ¬nh
model_1.compile(optimizer='adam',
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])

# 5. HuбєҐn luyб»‡n
model_1.fit(X_train_c10, y_train_c10, epochs=10, validation_split=0.2, batch_size=64)

# 6. ДђГЎnh giГЎ
loss, acc = model_1.evaluate(X_test_c10, y_test_c10)
print(f"Дђб»™ chГ­nh xГЎc trГЄn tбє­p Test CIFAR10: {acc:.4f}")
```

BГЂI 2

```python
# BГ i 2: ANN cho bб»™ dб»Ї liб»‡u MNIST (Chб»Ї viбєїt tay sб»‘)
print("--- BГ i 2: MNIST ---")

# 1. TбєЈi dб»Ї liб»‡u (GiбєЈ Д‘б»‹nh dГ№ng bбєЈn Keras do cбєҐu trГєc giб»‘ng hб»‡ dataset GV cung cбєҐp)
mnist = tf.keras.datasets.mnist
(X_train_mn, y_train_mn), (X_test_mn, y_test_mn) = mnist.load_data()

# 2. Tiб»Ѓn xб»­ lГЅ (Chuбє©n hГіa vб»Ѓ 0-1)
X_train_mn = X_train_mn / 255.0
X_test_mn = X_test_mn / 255.0

# 3. Khб»џi tбєЎo mГґ hГ¬nh
model_2 = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# 4. BiГЄn dб»‹ch mГґ hГ¬nh
model_2.compile(optimizer='adam',
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])

# 5. HuбєҐn luyб»‡n
model_2.fit(X_train_mn, y_train_mn, epochs=10, validation_split=0.1)

# 6. ДђГЎnh giГЎ
loss_mn, acc_mn = model_2.evaluate(X_test_mn, y_test_mn)
print(f"Дђб»™ chГ­nh xГЎc trГЄn tбє­p Test MNIST: {acc_mn:.4f}")
```

BГЂI 3

```python
# BГ i 3: Viбєїt chЖ°ЖЎng trГ¬nh cГ i Д‘бє·t ANN Д‘б»ѓ nhбє­n dбєЎng бєЈnh Cat hoбє·c Dog. (Sб»­ dб»Ґng dб»Ї liб»‡u giбєЈ)
print("--- BГ i 3: Cat vs Dog ---")

import numpy as np
from sklearn.model_selection import train_test_split
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Flatten, Dropout

num_samples = 100
X_cd = np.random.randint(0, 256, size=(num_samples, 200, 200, 3))
y_cd = np.random.randint(0, 2, size=num_samples)

X_cd = X_cd / 255.0

X_train_cd, X_test_cd, y_train_cd, y_test_cd = train_test_split(X_cd, y_cd, test_size=0.2, random_state=42)

model_3 = Sequential([
    Flatten(input_shape=(200, 200, 3)),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])

model_3.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model_3.fit(X_train_cd, y_train_cd, epochs=10, validation_data=(X_test_cd, y_test_cd), batch_size=16)

loss, acc = model_3.evaluate(X_test_cd, y_test_cd, verbose=0)
print(f"Дђб»™ chГ­nh xГЎc trГЄn tбє­p Test: {acc:.4f}")
```

BГЂI 4

```python
# BГ i 4: Viбєїt chЖ°ЖЎng trГ¬nh cГ i Д‘бє·t ANN Д‘б»ѓ dб»± bГЎo thu nhбє­p mб»™t ngЖ°б»ќi > 50K/nДѓm hay <= 50K/nДѓm
print("\n--- BГ i 4: Adult Income ---")

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from keras.models import Sequential
from keras.layers import Dense, Dropout

n_records = 500
np.random.seed(42)

data = {
    'age': np.random.randint(18, 70, n_records),
    'workclass': np.random.choice(['Private', 'Self-emp', 'Gov'], n_records),
    'education': np.random.choice(['Bachelors', 'HS-grad', 'Masters', 'PhD'], n_records),
    'hours_per_week': np.random.randint(20, 60, n_records),
    'income': np.random.choice(['<=50K', '>50K'], n_records)
}
adult_df = pd.DataFrame(data)

X_adult = adult_df.drop('income', axis=1)
y_adult = adult_df['income']

le = LabelEncoder()
y_adult = le.fit_transform(y_adult)

X_adult = pd.get_dummies(X_adult, drop_first=True)

X_train_ad, X_test_ad, y_train_ad, y_test_ad = train_test_split(X_adult, y_adult, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train_ad = scaler.fit_transform(X_train_ad)
X_test_ad = scaler.transform(X_test_ad)

model_4 = Sequential([
    Dense(64, activation='relu', input_dim=X_train_ad.shape[1]),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

model_4.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model_4.fit(X_train_ad, y_train_ad, epochs=15, validation_split=0.2, batch_size=32, verbose=0)

loss, acc = model_4.evaluate(X_test_ad, y_test_ad, verbose=0)
print(f"Accuracy trГЄn tбє­p test Adult: {acc:.4f}")
```

```python
# BГ i 5: Viбєїt chЖ°ЖЎng trГ¬nh cГ i Д‘бє·t ANN Д‘б»ѓ dб»± bГЎo viб»‡c Д‘ГЎnh giГЎ chбєҐt lЖ°б»Јng xe Гґ tГґ (nhГЈn cuб»‘i) tб»« bб»™ dб»Ї liб»‡u car. (Sб»­ dб»Ґng dб»Ї liб»‡u giбєЈ)
print("\n--- BГ i 5: Car Evaluation ---")

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.models import Sequential
from keras.layers import Dense

n_cars = 600
car_data = {
    'buying': np.random.choice(['vhigh', 'high', 'med', 'low'], n_cars),
    'maint': np.random.choice(['vhigh', 'high', 'med', 'low'], n_cars),
    'doors': np.random.choice(['2', '3', '4', '5more'], n_cars),
    'safety': np.random.choice(['low', 'med', 'high'], n_cars),
    'class': np.random.choice(['unacc', 'acc', 'good', 'vgood'], n_cars)
}
car_df = pd.DataFrame(car_data)

X_car = car_df.drop('class', axis=1)
y_car = car_df['class']

le_car = LabelEncoder()
y_car = le_car.fit_transform(y_car)

X_car = pd.get_dummies(X_car)

X_train_car, X_test_car, y_train_car, y_test_car = train_test_split(X_car, y_car, test_size=0.2, random_state=42)
num_classes = len(np.unique(y_car))

model_5 = Sequential([
    Dense(64, activation='relu', input_dim=X_train_car.shape[1]),
    Dense(32, activation='relu'),
    Dense(num_classes, activation='softmax')
])

model_5.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model_5.fit(X_train_car, y_train_car, epochs=20, validation_split=0.2, verbose=0)

loss, acc = model_5.evaluate(X_test_car, y_test_car, verbose=0)
print(f"Accuracy trГЄn tбє­p test Car: {acc:.4f}")

model_5.save('car_model_mock.h5')
```
