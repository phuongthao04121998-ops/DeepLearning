# **BT11**

```
x11 = tf.random.uniform([3, 2], minval=5, maxval=10)
y11 = tf.random.uniform([2, 3], minval=15, maxval=20)

# Lấy giá trị đường chéo của x
diag_x = tf.linalg.diag_part(x11)

# Thực hiện x + y (y đảo hàng thành cột)
result11 = x11 + tf.transpose(y11)

print("Đường chéo x:", diag_x.numpy())
print("Kết quả x + y_transpose:\n", result11.numpy())
```

# **BT12**

```
import tensorflow as tf

@tf.function # Chuyển hàm vào trong Graph
def compute_tensors(x, y):
    add = tf.add(x, y)
    sub = tf.subtract(x, y)
    mul = tf.multiply(x, y)
    div = tf.divide(x, y)
    return add, sub, mul, div

# Tạo 2 tensor ngẫu nhiên kích thước (3, 2, 2)
x = tf.random.uniform([3, 2, 2])
y = tf.random.uniform([3, 2, 2])

add, sub, mul, div = compute_tensors(x, y)
print("Bài 12 - Kết quả cộng (add):\n", add.numpy())
```

# **BT13-14**

```
class BasicMathModel(tf.Module):
    @tf.function
    def add(self, x, y):
        return x + y

    @tf.function
    def subtract(self, x, y):
        return x - y

    @tf.function
    def multiply(self, x, y):
        return x * y

model = BasicMathModel()
x = tf.constant([10.0, 20.0])
y = tf.constant([2.0, 5.0])

print("Bài 13 - Tổng:", model.add(x, y).numpy())
print("Bài 14 - Hiệu:", model.subtract(x, y).numpy())
print("Bài 14 - Tích:", model.multiply(x, y).numpy())
```

# **BT16**

```
@tf.function
def tinh_dien_tich_tp(a, b, c):
    s_xung_quanh = (a + b) * c * 2
    s_toan_phan = (a * b) * 2 + s_xung_quanh
    return s_toan_phan

a, b, c = tf.constant(2.0), tf.constant(3.0), tf.constant(4.0)
print(f"Bài 16 - Diện tích toàn phần: {tinh_dien_tich_tp(a, b, c).numpy()}")
```

# **BT17**

```
@tf.function
def check_parallel(AD_val, BC_val, AB_val, CD_val, is_parallel_input):
    # Kiểm tra điều kiện vế phải: AD = BC và AB = CD
    cond = tf.logical_and(tf.equal(AD_val, BC_val), tf.equal(AB_val, CD_val))
    # Trả về True nếu cả hai vế tương đương nhau
    return tf.equal(is_parallel_input, cond)

# Ví dụ: AD=5, BC=5, AB=3, CD=3 và giả định AD // BC là True
print("Bài 17 - Điều kiện đúng hay không:", check_parallel(5, 5, 3, 3, True).numpy())
```

# **BT18**

```
@tf.function
def check_equal_bases(AB_val, CD_val, AD_val, BC_val, is_parallel):
    # Điều kiện vế phải: AD = BC và AD // BC (is_parallel là True)
    cond = tf.logical_and(tf.equal(AD_val, BC_val), is_parallel)
    # So sánh với vế trái AB = CD
    return tf.equal(tf.equal(AB_val, CD_val), cond)

# Ví dụ
print("Bài 18 - Kiểm tra tính đúng đắn:", check_equal_bases(4, 4, 6, 6, True).numpy())
```

# **BT19**

```
import tensorflow as tf

@tf.function
def chung_minh_duong_trung_binh(AM, MB, AN, NC, MN, BC, is_parallel_MN_BC):
    """
    Xây dựng model chứng minh NA = NC dựa trên các điều kiện đầu vào.
    is_parallel_MN_BC: Tensor kiểu Boolean (True nếu MN // BC)
    """

    # 1. Kiểm tra điều kiện M là trung điểm AB (AM = MB)
    is_M_midpoint = tf.equal(AM, MB)

    # 2. Kiểm tra điều kiện MN là đường trung bình
    condition_to_prove = tf.logical_and(tf.equal(MN, MB), is_parallel_MN_BC)

    # 3. Kết luận NA = NC nếu thỏa mãn điều kiện trên
    is_proven = tf.logical_and(condition_to_prove, tf.equal(AN, NC))

    return is_proven

# --- Ví dụ ---
# Giả sử: AM=3, MB=3 (M là trung điểm), AN=4, NC=4, MN=3, BC=6 và MN // BC
AM = tf.constant(3.0)
MB = tf.constant(3.0)
AN = tf.constant(4.0)
NC = tf.constant(4.0)
MN = tf.constant(3.0)
BC = tf.constant(6.0)
is_parallel = tf.constant(True)

result = chung_minh_duong_trung_binh(AM, MB, AN, NC, MN, BC, is_parallel)

print("Kết quả chứng minh NA = NC:")
if result.numpy():
    print("=> Chứng minh THÀNH CÔNG: Điều kiện thỏa mãn để N là trung điểm AC.")
else:
    print("=> Chứng minh THẤT BẠI: Các điều kiện đầu vào không khớp.")
```
