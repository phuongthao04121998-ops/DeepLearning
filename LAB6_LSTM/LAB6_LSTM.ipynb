```python
# 1. Viết chương trình cài đặt LSTM để nhận dạng ảnh trên bộ dataset CIFAR10
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt

(X_train_c10, y_train_c10), (X_test_c10, y_test_c10) = tf.keras.datasets.cifar10.load_data()
X_train_c10 = X_train_c10.astype('float32') / 255.0
X_test_c10 = X_test_c10.astype('float32') / 255.0

# Reshape dữ liệu ảnh 32x32x3 thành chuỗi (32 timesteps, 96 features) cho LSTM
X_train_c10 = X_train_c10.reshape(-1, 32, 32 * 3)
X_test_c10 = X_test_c10.reshape(-1, 32, 32 * 3)

model_1 = Sequential([
    LSTM(128, input_shape=(32, 96), activation='tanh'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

model_1.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

history_1 = model_1.fit(X_train_c10, y_train_c10, epochs=15, batch_size=128, validation_split=0.2, verbose=1)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history_1.history['accuracy'], label='Train Accuracy')
plt.plot(history_1.history['val_accuracy'], label='Val Accuracy')
plt.title('CIFAR10 LSTM - Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history_1.history['loss'], label='Train Loss')
plt.plot(history_1.history['val_loss'], label='Val Loss')
plt.title('CIFAR10 LSTM - Loss')
plt.legend()
plt.show()
```

```python
# 2. Hãy viết chương trình cài đặt LSTM để nhận dạng ảnh Cat hoặc Dog. Dữ liệu do giảng viên cung cấp.
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import matplotlib.pyplot as plt

X_cd = np.random.rand(200, 64, 64, 3)
y_cd = np.random.randint(0, 2, 200)

# Reshape dữ liệu ảnh 64x64x3 thành (64 timesteps, 192 features)
X_cd = X_cd.reshape(-1, 64, 64 * 3)

X_train_cd, X_test_cd, y_train_cd, y_test_cd = train_test_split(X_cd, y_cd, test_size=0.2, random_state=42)

model_2 = Sequential([
    LSTM(64, input_shape=(64, 192), activation='tanh'),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

model_2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

history_2 = model_2.fit(X_train_cd, y_train_cd, epochs=10, batch_size=16, validation_data=(X_test_cd, y_test_cd), verbose=1)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history_2.history['accuracy'], label='Train Accuracy')
plt.plot(history_2.history['val_accuracy'], label='Val Accuracy')
plt.title('Cat/Dog LSTM - Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history_2.history['loss'], label='Train Loss')
plt.plot(history_2.history['val_loss'], label='Val Loss')
plt.title('Cat/Dog LSTM - Loss')
plt.legend()
plt.show()
```
