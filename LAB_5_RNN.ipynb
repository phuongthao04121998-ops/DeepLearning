```python
# 1. Cài đặt RNN để dự báo giá nhà (price)
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, SimpleRNN
import matplotlib.pyplot as plt

data = np.cumsum(np.random.normal(10, 5, 500)).reshape(-1, 1)

time_step = 10
X, y = [], []
for i in range(len(data) - time_step):
    X.append(data[i:(i + time_step), 0])
    y.append(data[i + time_step, 0])
X, y = np.array(X), np.array(y)
X = np.reshape(X, (X.shape[0], X.shape[1], 1))

train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

model_1 = Sequential()
model_1.add(SimpleRNN(32, input_shape=(time_step, 1), activation='relu'))
model_1.add(Dense(1))
model_1.compile(optimizer='adam', loss='mse')

history_1 = model_1.fit(X_train, y_train, epochs=20, batch_size=16, validation_data=(X_test, y_test), verbose=1)

test_predict = model_1.predict(X_test)

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history_1.history['loss'], label='Train Loss')
plt.plot(history_1.history['val_loss'], label='Val Loss')
plt.title('House Price - Model Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(y_test, label='Actual Price')
plt.plot(test_predict, label='Predicted Price')
plt.title('House Price - Actual vs Predicted')
plt.legend()
plt.show()
```
